# Executing SQL

## Query Tables in Database

````python
%%sql 
SELECT t.TABLE_NAME, t.TABLE_COMMENT
FROM information_schema.TABLES t
WHERE t.TABLE_SCHEMA='sysbench'
ORDER BY t.TABLE_NAME;
````

## Query Columns in Table

````python
%%sql 
SELECT ORDINAL_POSITION, COLUMN_NAME, COLUMN_TYPE, IS_NULLABLE, COLUMN_DEFAULT, COLUMN_COMMENT
FROM information_schema.COLUMNS
WHERE TABLE_SCHEMA='sysbench' 
AND TABLE_NAME='customer1'
ORDER BY ORDINAL_POSITION;
````

## Query Table Data

````python
%%sql 
SELECT *
FROM sysbench.customer1
LIMIT 10;
````
