# Environment

```python
import requests
import json
import base64
import pprint

headers = {
    'Content-Type': 'application/json',
    # Basic Authorization
    'Authorization': 'Basic ' + base64.b64encode('elastic:es+apm'.encode()).decode()
}

HOST = 'localhost'
URL_PREFIX = 'http://{host}:9200'.format(host=HOST)

# Requests
# https://docs.python-requests.org/en/latest/api


def print_request(pr):
    body = pr.body
    if isinstance(body, str):
        body = body.encode('utf-8').decode('unicode_escape')
    if body != None and len(body) > 1000:
        body = body[:100] + '...'
    print("""=== >
{method} {url}
{headers}
{body}""".format(method=pr.method, url=pr.url, headers=pr.headers, body=body))


def print_response(r):
    print("====== Response\n", r.status_code)
    print_request(r.request)
    print("=== <")
    print(r.headers)
    if 'content-type' in r.headers and 'application/json' in r.headers['content-type']:
        pprint.pprint(r.json())
    else:
        if len(r.content) > 1000:
            print(r.content[:100], '...')
        else:
            print(r.content)


url = URL_PREFIX+"/_cat"
r = requests.get(url=url, headers=headers, params={})
print(r.text)
```
